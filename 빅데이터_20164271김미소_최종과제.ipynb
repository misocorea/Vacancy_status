```python
import pandas as pd
import numpy as np

import platform
import matplotlib.pyplot as plt

%matplotlib inline

path = "c:/Windows/Fonts/malgun.ttf"
from matplotlib import font_manager, rc
if platform.system() == 'Darwin' :
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system... sorry~~~~')
    
plt.rcParams['axes.unicode_minus'] = False
```

```python
empty_Seoul = pd.read_excel('../data/빈집Report.xls', encoding = 'utf-8')
empty_Seoul.head()
```

```python
empty_Seoul = pd.read_excel('../data/빈집Report.xls', header = 1,encoding = 'utf-8')
empty_Seoul.head()
```

```python
empty_Seoul.drop(empty_Seoul.index[[0]], inplace=True)
empty_Seoul.head()
```

```python
del empty_Seoul['기간']
empty_Seoul.head()
```

```python
empty_Seoul.set_index('구분', inplace=True)
empty_Seoul.head()
```

```python
empty_Seoul['계'].plot(kind='barh',grid=True, figsize=(10,10))
plt.show()
```

```python
empty_Seoul['계'].sort_values().plot(kind='barh', grid=True, figsize=(10,10))
plt.show()
```

```python
import json
geo_path = '../data/02. skorea_municipalities_geo_simple.json'
geo_str = json.load(open(geo_path, encoding='utf-8'))
```

```python
import folium
```

```python
map = folium.Map(location=[37.5502, 126.982], zoom_start=11,
                tiles='Stamen Toner')
map.choropleth(geo_data = geo_str, 
               data = empty_Seoul['계'], 
               columns = [empty_Seoul.index, 
                          empty_Seoul['계']],
               fill_color ='PuRd', #Rurd, YlGnBu
               key_on = 'feature.id')
map
```
